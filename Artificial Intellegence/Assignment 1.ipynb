```python
# Question 1

def prime(n):
    is_prime = False
    if n <= 1:
        print("The number is negative!")
    else:
        is_prime = True
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                is_prime = False
                break
    if is_prime:
        print("The number is prime.")
    else:
        print("The number is not prime.")

# Example of the output:
prime(7)
prime(-2)
prime(9)
prime(10)
```

```python
# Question 2:

def sum_digits(number):
    total = 0
    for digit_char in str(number):
        total += int(digit_char)
    return total


# Example of the output:
num = 12345
result = sum_digits(num)

print(f"The number is: {num}")
print(f"The sum of its digits is: {result}")
```

```python
# Question 4

my_list = ['madam', 'hello', 'level', 'world', 'radar']
palindromes = []
non_palindromes = []

for word in my_list:
    if word == word[::-1]:
        palindromes.append(word)
    else:
        non_palindromes.append(word)

print("Palindrome Words:")
for p in palindromes:
    print(p)

print("\nNon Palindrome Words:")
for np in non_palindromes:
    print(f"'{np}'")
```

```python
# Question 5

sentence = "Hello world, this is a Python program."

word_count = 0
i = 0
in_word = False
length = len(sentence)

while i < length:
    char = sentence[i]
    if not char.isspace():
        if not in_word:
            word_count += 1
            in_word = True
    else:
        in_word = False
    i += 1

print(f"The sentence is: '{sentence}'")
print(f"The number of words is: {word_count}")
```

```python
# Question 6

num1 = 48
num2 = 18

a = num1
b = num2

while b:
    a, b = b, a % b

gcd = a

print(f"num1 = {num1}")
print(f"num2 = {num2}")
print(f"The GCD of the given numbers is: {gcd}")
```

```python
# Question 7

a = num1
b = num2

while b:
    a, b = b, a % b

gcd = a

lcm = (num1 * num2) // gcd

print(f"num1 = {num1}")
print(f"num2 = {num2}")
print(f"The LCM of {num1} and {num2} is: {lcm}")
```

```python
# Task 1

try:
    salary_input = input("Please enter your current salary: ")
    salary = float(salary_input)

    raise_amount = salary * 0.20
    new_salary = salary + raise_amount

    print(f"Your current salary is: ${salary:,.2f}")
    print(f"Your 20% raise is: ${raise_amount:,.2f}")
    print(f"Your new salary will be: ${new_salary:,.2f}")

except ValueError:
    print("Invalid input. Please enter a valid number for your salary.")
```

```python
# Task 2

try:
    num1 = float(input("Enter the first number: "))
    operator = input("Enter an operator (+, -, *, /): ")
    num2 = float(input("Enter the second number: "))

    result = 0

    if operator == '+':
        result = num1 + num2
    elif operator == '-':
        result = num1 - num2
    elif operator == '*':
        result = num1 * num2
    elif operator == '/':
        if num2 == 0:
            print("Error: Cannot divide by zero.")
            result = "undefined"
        else:
            result = num1 / num2
    else:
        print("Error: Invalid operator entered.")
        result = "invalid"

    if result != "undefined" and result != "invalid":
        print(f"The result of {num1} {operator} {num2} is: {result}")

except ValueError:
    print("Invalid input. Please enter valid numbers.")
```

```python
# Task 3

try:
    num_students = int(input("Enter the number of students: "))

    if num_students <= 0:
        print("Please enter a positive number of students.")
    else:
        scores = []
        for i in range(num_students):
            while True:
                try:
                    score = float(input(f"Enter score for student {i + 1}: "))
                    scores.append(score)
                    break
                except ValueError:
                    print("Invalid score. Please enter a number.")

        average_grade = sum(scores) / len(scores)

        print(f"\nAll scores: {scores}")
        print(f"Average grade: {average_grade:.2f}")

        if average_grade >= 60:
            print("Result: pass")
        else:
            print("Result: fail")

except ValueError:
    print("Invalid input. Please enter a valid integer for the number of students.")
```
